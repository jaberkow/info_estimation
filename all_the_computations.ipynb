```python
info_df = pd.read_pickle('pop_parameters.pkl')
```

```python
num_sets = len(info_df)
```

```python
results_df = info_df[info_df.columns[:4]]
```

```python
info_df.columns
```

```python
results_df["r binary"] = ""
results_df["spike count"]= ""
results_df["M binarized exact"] = ""
results_df["M binarized binned cartesian"] = ""
results_df["M unbinarized binned cartesian"] = ""
results_df["M binarized binned log-polar"] = ""
results_df["M unbinarized binned log-polar"] = ""

results_df["U binarized exact"] = ""
results_df["U binarized binned cartesian"] = ""
results_df["U unbinarized binned cartesian"] = ""
results_df["U binarized binned log-polar"] = ""
results_df["U unbinarized binned log-polar"] = ""

results_df["U binarized exact bad theta"] = ""
```

```python
results_df
```

```python
#13 computations per subset
for i in range(num_sets):
    raw_data_block_name = info_df['Filename'].iloc[i]
    raw_data_block = np.load(raw_data_block_name)
    
    beta_vec = info_df['Beta Vectors'].iloc[i]
    stand_in_beta = np.ones_like(beta_vec)
    theta_vec = info_df['Theta Vectors (radians)'].iloc[i]
    coarse_theta_vec = info_df['Worst Case Theta (radians)'].iloc[i]
    
    w = dcu.get_w(stand_in_beta,theta_vec)
    w_tilde = dcu.get_w(beta_vec,theta_vec)
    w_coarse = dcu.get_w(stand_in_beta,coarse_theta_vec)
    
    to_seconds_for_integer = 1.0/0.03
    r_max = np.amax(raw_data_block)
    to_seconds_for_binary = float(r_max)*to_seconds_for_integer
    
    raw_binary_block = dcu.block_binarizer(raw_data_block)
    signed_raw_binary_block = 2.0*raw_binary_block - 1.0
    
    #binary patterns information
    binary_inds = dcu.get_binary_inds(raw_binary_block)
    results_df["r binary"].iloc[i] = dcu.compute_information(binary_inds,to_seconds_for_binary)
    
    #spike count information
    spike_count_inds = dcu.get_spike_count_inds(raw_binary_block)
    results_df["spike count"].iloc[i] = dcu.compute_information(spike_count_inds,to_seconds_for_binary)
    
    #Exact M after binarization information
    M_binary_exact_inds = dcu.get_exact_vector_inds(binary_inds,w_tilde)
    results_df["M binarized exact"].iloc[i] = dcu.compute_information(M_binary_exact_inds,to_seconds_for_binary)
    
    #M binned cartesian after binarization information
    M_binarized_binned_cart_inds = dcu.get_2d_hist_inds(signed_raw_binary_block,w_tilde,lp_transform=False)
    results_df["M binarized binned cartesian"].iloc[i] = dcu.compute_information(M_binarized_binned_cart_inds,to_seconds_for_binary)
    
    #M binned cartesian no binarization information
    M_unbinarized_binned_cart_inds = dcu.get_2d_hist_inds(raw_data_block,w_tilde,lp_transform=False)
    results_df["M unbinarized binned cartesian"].iloc[i] = dcu.compute_information(M_unbinarized_binned_cart_inds,to_seconds_for_integer)
    
    #M binned log polar after binarization information
    M_binarized_binned_lp_inds = dcu.get_2d_hist_inds(signed_raw_binary_block,w_tilde,lp_transform=True)
    results_df["M binarized binned log-polar"].iloc[i] = dcu.compute_information(M_binarized_binned_lp_inds,to_seconds_for_binary)
    
    #M binned log polar no binarization
    M_unbinarized_binned_lp_inds = dcu.get_2d_hist_inds(raw_data_block,w_tilde,lp_transform=True)
    results_df["M unbinarized binned log-polar"].iloc[i] = dcu.compute_information(M_unbinarized_binned_lp_inds,to_seconds_for_integer)
    
    #Exact U after binarization information
    U_binary_exact_inds = dcu.get_exact_vector_inds(binary_inds,w)
    results_df["U binarized exact"].iloc[i] = dcu.compute_information(U_binary_exact_inds,to_seconds_for_binary)
    
    #U binned cartesian after binarization information
    U_binarized_binned_cart_inds = dcu.get_2d_hist_inds(signed_raw_binary_block,w,lp_transform=False)
    results_df["U binarized binned cartesian"].iloc[i] = dcu.compute_information(U_binarized_binned_cart_inds,to_seconds_for_binary)
    
    #U binned cartesian no binarization information
    U_unbinarized_binned_cart_inds = dcu.get_2d_hist_inds(raw_data_block,w,lp_transform=False)
    results_df["U unbinarized binned cartesian"].iloc[i] = dcu.compute_information(U_unbinarized_binned_cart_inds,to_seconds_for_integer)
    
    #U binned log polar after binarization information
    U_binarized_binned_lp_inds = dcu.get_2d_hist_inds(signed_raw_binary_block,w,lp_transform=True)
    results_df["U binarized binned log-polar"].iloc[i] = dcu.compute_information(U_binarized_binned_lp_inds,to_seconds_for_binary)
    
    #U binned log polar no binarization
    U_unbinarized_binned_lp_inds = dcu.get_2d_hist_inds(raw_data_block,w,lp_transform=True)
    results_df["U unbinarized binned log-polar"].iloc[i] = dcu.compute_information(U_unbinarized_binned_lp_inds,to_seconds_for_integer)
    
    #U binarized exact with coarse theta
    U_binary_exact_coarse_inds = dcu.get_exact_vector_inds(binary_inds,w_coarse)
    results_df["U binarized exact bad theta"].iloc[i] = dcu.compute_information(U_binary_exact_coarse_inds,to_seconds_for_binary)
    
    print "finished " + str(info_df["subset"].iloc[i]) + ", {} out of {}".format(i+1,num_sets)
```
